````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
# print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
Movie_url  =[]
movies_link = soup.find_all("a",class_="title-list-grid__item--link")
for link in movies_link:
    movie_url = "https://www.justwatch.com"+link["href"]
    Movie_url.append(movie_url)
print(Movie_url)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
Movie_title = []
for i in Movie_url:
    res = requests.get(i)
    soup = BeautifulSoup(res.content, "html.parser")
    title_block = soup.find(class_="title-block")
    try:
        if title_block:
            title = title_block.h1.text.strip()
            Movie_title.append(title)
            # print(title)
        else:
            # print("NA")
            Movie_title.append("NA")
    except Exception as e:
        # print("Error processing URL",i)
        Movie_title.append("NA")

print(Movie_title)
````

````python
len(Movie_title)
````

## **Scrapping release Year**

````python
# # Write Your Code here
Release_year = []
for i in Movie_url:
    res = requests.get(i)
    soup = BeautifulSoup(res.content, "html.parser")
    year_block = soup.find(class_="title-block")

    try:
        if year_block:
            year = year_block.span.text.strip().replace('(', '').replace(')', '')
            Release_year.append(year)
            # print(year)
        else:
            # print("NA")
            Release_year.append("NA")
    except Exception as e:
        # print("Error processing URL",i)
        Release_year.append("NA")

print(Release_year)
````

````python
print(len(Release_year))
print(Release_year.count('NA'))
````

## **Scrapping Genres**

````python
# Write Your Code here
Genre = []

for i in Movie_url:
    res = requests.get(i)
    soup = BeautifulSoup(res.content, "html.parser")
    genres_heading = soup.find('h3', class_='detail-infos__subheading', string='Genres')
    try:
        if genres_heading:
            genres_element = genres_heading.find_next_sibling('div', class_='detail-infos__value')
            genres = genres_element.text.strip()
            Genre.append(genres)
            # print(genres)
        else:
            Genre.append("NA")
            # print("NA")
    except Exception as e:
        # print("Error processing URL",i)
        Genre.append("NA")
print(Genre)
````

````python
print(Genre.count("NA"))
len(Genre)
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
IMDB_rating =[]
for i in range(100):
    res = requests.get(Movie_url[i])
    soup = BeautifulSoup(res.content, "html.parser")
    imdb_block = soup.find('img', alt='IMDB')
    if imdb_block:
      rating_span = imdb_block.find_next('span')
      if rating_span:
        imdb_rating = rating_span.text.strip().split()[0]
        IMDB_rating.append(imdb_rating)
        # print(imdb_rating)
      else:
        IMDB_rating.append("NA")
        # print("NA")
    else:
        IMDB_rating.append("NA")
        # print("NA")

print(IMDB_rating)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
Runtime = []
for i in range(100):
    try:
        res = requests.get( Movie_url[i])
        res.raise_for_status()  # Raise an exception for 4xx and 5xx status codes
        soup = BeautifulSoup(res.text, "html.parser")

        # Find the element containing IMDb rating information
        detail_infos_divs = soup.find_all('div', class_='detail-infos')

        # Iterate through each div and check if it contains the text 'Runtime'
        runtime_found = False
        for div in detail_infos_divs:
            if div.find('h3', class_='detail-infos__subheading').text.strip() == 'Runtime':
                # Extract the text content of the div with class 'detail-infos__value'
                runtime = div.find('div', class_='detail-infos__value').text.strip()
                # print(runtime)
                if runtime:
                    Runtime.append(runtime)
                else:
                    Runtime.append("NA")
                runtime_found = True
                break  # Stop searching after finding the first runtime information
        if not runtime_found:
            Runtime.append("NA")
    except Exception as e:
#         print(f"Error processing URL {url1}: {e}")
        Runtime.append("NA")

print(Runtime)
````

````python
# Write Your Code here
Streaming_services = []
for i in range(100):
    res = requests.get(Movie_url[i])
    soup = BeautifulSoup(res.content, "html.parser")

    try:
        # Look for elements with class 'offer'
        offer_elements = soup.find_all(class_='offer')

        # Initialize a variable to store the service name
        service_name = "NA"

        # Iterate over offer elements to find the service name
        for offer_element in offer_elements:
            img_element = offer_element.find('img', class_='offer__icon')
            if img_element:
                # Get the value of the 'alt' attribute, which contains the service name
                service_name = img_element.get('alt')
                break  # Stop searching after finding the first service name

        Streaming_services.append(service_name)
        print(service_name)

    except Exception as e:
        # print(f"Error processing URL {Movie_url[i]}: {e}")
        Streaming_services.append("NA")

print(Streaming_services)
````
